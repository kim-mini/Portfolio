```python
from tensorflow.keras.datasets import imdb
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud
from wordcloud import STOPWORDS
import os
import pandas as pd
from PIL import Image
```

```python
path = '../natural_language_processing/IMDB_Dataset.csv'

if os.path.isfile(path):
    pass
else:
    print("Not found '{}'".format(path))
```

```python
df = pd.read_csv(path)
```

```python
df.head()
```

```python
for i in range(len(df)):
    text += df['review'][i]
wordcloud = WordCloud().generate(text)
```

```python
# word cloud 
plt.figure(figsize=(12,12))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

# Alice

```python
# text  
text = open('./alice.txt').read()
```

```python
alice_mask = np.array(Image.open('./alice_mask.png'))

#   text  
# stopwords = set(STOPWORDS)
# stopwords.add('said')
```

```python
# mask 
plt.figure(figsize=(12,12))
plt.imshow(alice_mask,cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(background_color='white', max_words=2000, mask=alice_mask)
wc = wc.generate(text)
```

```python
# mask text 
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
alice_mask2 = np.array(Image.open('./alice_col__mask.png'))
```

```python
wc2 = WordCloud(background_color='white', max_words=2000, mask=alice_mask2)
wc2.generate(text)
```

```python
from wordcloud import ImageColorGenerator
image_colors = ImageColorGenerator(alice_mask2)
```

```python
# mask 
plt.figure(figsize=(12,12))
plt.imshow(alice_mask2,cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# mask 
plt.figure(figsize=(12,12))
plt.imshow(wc2, interpolation='bilinear')
plt.axis('off')
plt.show()
```

##   

```python
plt.figure(figsize=(12,12))
plt.imshow(wc2.recolor(color_func=image_colors), interpolation='bilinear')
plt.axis('off')
plt.show()
```
